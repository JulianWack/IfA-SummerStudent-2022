### Create halos
In Getting Started -> Catalogs, created halos as uniform catalog of points and gave them mass, lending to the interpretation of being a halo:
```
cat = UniformCatalog(nbar=100, BoxSize=1.0, seed=42)
cat['Mass'] = 10**(cat.rng.uniform(12,15))
halos = HaloCatalog(cat, cosmo=cosmology.Planck15, redshift=0., mdef='vir', position='Position', velocity='Velocity', mass='Mass')
hod = halos.populate(Zheng07Model)
```

Here the process is: Generate a log-normal catalog of objects, where ever these produce a group/ lump together, we define the group of these objects as a halo. Then populate the halos with galaxies. The grouping is done via the friends-of-friends (FOF) algorithm:

FOF (or also DBSCAN) is an algorithm that identifies groups of particles. This is done by considering the distance between two particles relative to the mean separation. Should this distance for a pair of galaxies be less than or equal to a user defined ```linking_length```, then the galaxies are considered to be directly linked (friends). A galaxy is indirectly linked to all the friends of its friends (friends of friends). This creates networks of linked particles which are called groups. A group is accepted when it contains at least ```nmin``` galaxies. 

````python
redshift = 0.55
cosmo = cosmology.Planck15
b1 = 2.0
Plin = cosmology.LinearPower(cosmo, redshift, transfer='EisensteinHu')
````

````python
# generate object catalog from log-normal densnsity field as seed for halos
seed_cat = LogNormalCatalog(Plin=Plin, nbar=3e-3, BoxSize=1380., Nmesh=256, bias=b1, seed=42)

# FOF algorithm on the seed catalog to generate a catalog of halos
fof = FOF(seed_cat, linking_length=0.2, nmin=5)
halos = fof.to_halos(cosmo=cosmo, redshift=redshift, particle_mass=1e12, mdef='vir')

# populate halos to create galaxy catalog: hod
from nbodykit.hod import Zheng07Model
hod = halos.populate(Zheng07Model, seed=42) # instructions in tutorial notebook outdated
````

````python
# extract subcatalogs of centrals and sattelites
cen_bool = hod['gal_type'] == 0
sat_bool = hod['gal_type'] == 1

cens = hod[cen_bool]
sats = hod[sat_bool]
````

### Real space power spectra
For halos, hod galaxies, and also for only centrals or only satellites. Can also find cross power between centrals and satellites 

````python
halo_power = FFTPower(halos, mode='1d', Nmesh=512)
gal_power = FFTPower(hod, mode='1d', Nmesh=256)
cen_power = FFTPower(cens, mode='1d', Nmesh=256)
sat_power = FFTPower(sats, mode='1d', Nmesh=256)

cen_sat_power = FFTPower(cens, second=sats, mode='1d', Nmesh=256)
````

````python
# plot galaxy auto power, centrals auto power, and sats auto power
labels = [r"$P^{gg}$", r"$P^{cc}$", r"$P^{ss}$"]
for i, r in enumerate([gal_power, cen_power, sat_power]):
    Pk = r.power
    plt.loglog(Pk['k'], Pk['power'].real-Pk.attrs['shotnoise'], label=labels[i])

# central-satellite power
Pk = cen_sat_power.power
plt.loglog(Pk['k'], Pk['power'].real, label=r"$P^{cs}$")

# the halo power
Phalo = halo_power.power
plt.loglog(Phalo['k'], Phalo['power'].real-Phalo.attrs['shotnoise'], c='k', label=r"$P^\mathrm{halo}$")

# add a legend and axis labels
plt.legend(loc=0, ncol=2)
plt.xlabel(r"$k$ [$h \ \mathrm{Mpc}^{-1}$]")
plt.ylabel(r"$P(k)$ [$h^{-3}\mathrm{Mpc}^3$]")
plt.xlim(0.01, 0.6)
plt.ylim(1e4, 5e6)
plt.show()
````

### Redshift space power spectra

````python
# converte HOD galaxy positions from real to redshift space
LOS = [0,0,1]
hod['RSDPosition'] = hod['Position'] + hod['VelocityOffset']*LOS

# find P(k, mu) for 5 mu bins as in LogNormalCatalog notebook
mesh = hod.to_mesh(position='RSDPosition', Nmesh=256, compensated=True)
rsd_gal_power = FFTPower(mesh, mode='2d', Nmu=5)
pkmu = rsd_gal_power.power
````

````python
# plot each mu bin
for i in range(pkmu.shape[1]):
    Pk = pkmu[:,i]
    label = r'$P^{gg}(\mu$=%.1f)' %pkmu.coords['mu'][i]
    plt.loglog(Pk['k'], Pk['power'].real - Pk.attrs['shotnoise'], label=label)

# the halo power
Phalo = halo_power.power
plt.loglog(Phalo['k'], Phalo['power'].real-Phalo.attrs['shotnoise'], c='k', label=r"$P^\mathrm{halo}$")

# add a legend and axis labels
plt.legend(loc=0, ncol=2)
plt.xlabel(r"$k$ [$h \ \mathrm{Mpc}^{-1}$]")
plt.ylabel(r"$P(k, \mu)$ [$h^{-3}\mathrm{Mpc}^3$]")
plt.xlim(0.01, 0.6)
plt.ylim(1e4, 2e6)
plt.show()
````
