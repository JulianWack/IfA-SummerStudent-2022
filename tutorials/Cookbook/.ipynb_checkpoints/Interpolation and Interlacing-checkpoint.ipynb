Compare 1D power spectrum as deduced from a log-normal mock catalog using different window kernels and applying (no) interlacing.

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
from nbodykit.lab import *
from nbodykit import style
from scipy.interpolate import InterpolatedUnivariateSpline as spline

import matplotlib.pyplot as plt
plt.style.use(style.notebook)
```

```python
redshift = 0.55
cosmo = cosmology.Planck15
Plin = cosmology.LinearPower(cosmo, redshift, transfer='EisensteinHu')

cat = LogNormalCatalog(Plin=Plin, nbar=3e-3, BoxSize=1380, Nmesh=256, bias=2.0, seed=42)
```

By using a large `Nmesh`, the effects of aliasing are reduced. With the high order interpolation window tsc, we regard the deduced power spectrum as the truth/target.

```python
r = FFTPower(cat.to_mesh(resampler='tsc', Nmesh=512, compensated=True), mode='1d')
truth = r.power
truth = spline(truth['k'], truth['power'].real - truth.attrs['shotnoise'])
```

### Compare different windows

```python
for window in ['cic', 'tsc', 'sym6', 'sym12', 'sym20', 'lanczos2', 'lanczos3']:

    print("computing power for window '%s:" %window)
    # only cic and tsc have build in compensation
    compensated = True if window in ['cic', 'tsc'] else False
    # convert catalog to a mesh with desired window and interlacing
    mesh = cat.to_mesh(Nmesh=256, resampler=window, compensated=compensated, interlaced=False)

    # compute the power
    %time r = FFTPower(mesh, mode='1d')
    print("-"*32)
    Pk = r.power

    plt.plot(Pk['k'], (Pk['power'].real - Pk.attrs['shotnoise']) / truth(Pk['k']), label=window)

plt.legend(loc=0, ncol=2, fontsize=16, frameon=False)
plt.xlabel(r"$k$ [$h \ \mathrm{Mpc}^{-1}$]")
plt.ylabel(r"$P(k) / P(k)^\mathrm{truth}$")
plt.xlim(0.01, 0.6)
plt.ylim(0.9, 1.1)
plt.show()
```

### Notes
Each window has some $k_{max}$ up to which the results are accurate.
Considering size of $k_{max}$ and run time advocates using `tsc` or `cis`.

Sym schemes take about an order of magnitude longer.

### Interlacing

```python
Nmesh = 256
for interlaced in [True, False]:
    for window in ['cic', 'tsc']:
        mesh = cat.to_mesh(Nmesh=Nmesh, resampler=window, compensated=True, interlaced=interlaced)

        print('Computing for %s; interlacing: %s' %(window, interlaced))
        %time r = FFTPower(mesh, mode='1d')
        print("-"*32)
        Pk = r.power

        # compare P(k) to the hi-resolution mesh P(k)
        label = 'interlaced=%s, window=%s' %(interlaced, window)
        plt.plot(Pk['k'], (Pk['power'].real - Pk.attrs['shotnoise']) / truth(Pk['k']), label=label)


# plot Nyquist frequency
BoxSize = 1380
k_ny = numpy.pi * Nmesh / BoxSize
plt.axvline(x=k_ny, c='k', label="Nyquist frequency for Nmesh=%d" %Nmesh)

# format the axes
plt.legend(loc=0, ncol=2, frameon=False)
plt.xlabel(r"$k$ [$h \ \mathrm{Mpc}^{-1}$]")
plt.ylabel(r"$P(k) / P(k)^\mathrm{truth}$")
plt.ylim(0.9, 1.2)
plt.show()
```

### Notes
Effect of aliasing significant for large $k$. Higher order interpolation gives better result and interlacing drastically increases accuracy for fixed window size.

## Recommendation
Use `tsc` and `interlacing=True`. To find a suitable `Nmesh` value, find power spectrum for small sample at high resolution and then try various `Nmesh` values in the same manner as above. Depending on the desired $k_{max}$ choose a value.
