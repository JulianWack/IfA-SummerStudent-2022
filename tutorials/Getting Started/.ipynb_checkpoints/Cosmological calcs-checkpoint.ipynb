```python
import numpy as np
import matplotlib.pyplot as plt
from nbodykit.lab import *

from nbodykit import style
plt.style.use(style.notebook)
```

## Power Spectra

```python
cosmo = cosmology.Planck15
linear_power_spec = cosmology.LinearPower(cosmo, redshift=0, transfer='CLASS')

ks = np.logspace(-3,0,100)
plt.loglog(ks, linear_power_spec(ks), c='k')

plt.xlabel(r"$k$ $[h \mathrm{Mpc}^{-1}]$")
plt.ylabel(r"$P$ $[h^{-3} \mathrm{Mpc}^{3}]$")
plt.show()
```

#### change redshift and power specturm normalisation i.e sigma8

```python
# original power
plt.loglog(ks, linear_power_spec(ks),
           label="$z={:.2f}, \sigma_8={:.2f}$".format(linear_power_spec.redshift, linear_power_spec.sigma8))

# update the redshift at which power is computed and sigma8 (the z=0 amplitude of matter fluctuations)
linear_power_spec.redshift = 0.55
linear_power_spec.sigma8 = 0.80
plt.loglog(ks, linear_power_spec(ks),
           label="$z={:.2f}, \sigma_8={:.2f}$".format(linear_power_spec.redshift, linear_power_spec.sigma8))

# format the axes
plt.legend()
plt.xlabel(r"$k$ $[h \mathrm{Mpc}^{-1}]$")
plt.ylabel(r"$P$ $[h^{-3} \mathrm{Mpc}^{3}]$")
plt.show()
```

#### comparing linear, non-linear, and Zel'dovich power spectra

```python
Plin = cosmology.LinearPower(cosmo, redshift=0, transfer='CLASS')
Pnl = cosmology.HalofitPower(cosmo, redshift=0)
Pzel = cosmology.ZeldovichPower(cosmo, redshift=0)

# plot each kind
plt.loglog(ks, Plin(ks), label='linear')
plt.loglog(ks, Pnl(ks), label='nonlinear')
plt.loglog(ks, Pzel(ks), label="Zel'dovich")

# format the axes
plt.legend()
plt.xlabel(r"$k$ $[h \mathrm{Mpc}^{-1}]$")
plt.ylabel(r"$P$ $[h^{-3} \mathrm{Mpc}^{3}]$")
plt.show()
```

## Correlation functions

```python
cor_func_lin = cosmology.CorrelationFunction(Plin)
cor_func_nl = cosmology.CorrelationFunction(Pnl)
cor_func_zel = cosmology.CorrelationFunction(Pzel)

# plot each kind
r = np.logspace(-1, np.log10(150), 1000)
plt.plot(r, r**2 * cor_func_lin(r), label='linear')
plt.plot(r, r**2 * cor_func_nl(r), label='nonlinear')
plt.plot(r, r**2 * cor_func_zel(r), label="Zel'dovich")

# format the axes
plt.legend()
plt.xlabel(r"$r$ $[h^{-1} \mathrm{Mpc}]$")
plt.ylabel(r"$r^2 \xi \ [h^{-2} \mathrm{Mpc}^2]$")
plt.show()
```
