# Description
Run a quick model fit to get initial guess of their mean value, which can then be used in the priors

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
import numpy as np
import matplotlib.pyplot as plt
from datetime import timedelta
import time
from scipy.optimize import minimize
from scipy.special import erf

import inv_cov_funcs as icf
import classylss
import fitsio
from nbodykit.lab import *
from nbodykit import style
plt.style.use(style.notebook)
```

```python
LOS = [0,0,1]
redshift = 0
BoxSize = 2000
cosmo_paras = classylss.load_ini('/home/jwack/main/Planck18_LCDM.ini')
cosmo = cosmology.cosmology.Cosmology.from_dict(cosmo_paras)
Plin = cosmology.LinearPower(cosmo, redshift, transfer='CLASS')

dk = 0.01
ells = [0,2]
```

```python
# load computed power spectra to deduce multipoles in each bin and P(k,mu) from data
k_full, shotnoise, n_ptile, Pk_ells_full = icf.load_power_data('/home/jwack/main/fit_linear_model_in_density_bins/', 
                                                               ells, get_data_Pkmus=False)
# for given BoxSize, k is NaN above 0.034
possible_kmax = k_full[k_full<=0.343][1:] # ignore first k bin 
```

```python
def chi2(theta, data_multipoles, k, C_inv, damped):
    '''Compute chi^2 for damped or undamped Kaiser model. 
    data_multipoles must be an array of shape (len(ells), len(k)). theta is the model parameter array.'''
    ells = [0,2]
    model_multipoles = np.empty((len(ells), len(k)))
    
    if damped:
        b1, beta, sigma = theta
        model_multipoles[0] = ( 1/(2*(k*sigma)**5) * (np.sqrt(2*np.pi)*erf(k*sigma/np.sqrt(2))*(3*beta**2+(k*sigma)**4+2*beta*(k*sigma)**2) + 
                                                    np.exp(-0.5*(k*sigma)**2)*(-2*beta*(beta+2)*(k*sigma)**3-6*beta**2*k*sigma) ) ) * b1**2 * Plin(k)
        model_multipoles[1] = ( -5/(4*(k*sigma)**7) * (np.sqrt(2*np.pi)*erf(k*sigma/np.sqrt(2))*(-45*beta**2+(k*sigma)**6+(2*beta-3)*(k*sigma)**4+3*(beta-6)*beta*(k*sigma)**2) + 
                                                    np.exp(-0.5*(k*sigma)**2)*((4*beta*(beta+2)+6)*(k*sigma)**5+12*beta*(2*beta+3)*(k*sigma)**3+90*beta**2*k*sigma) ) ) * b1**2 * Plin(k)
    else:
        b1, beta = theta
        model_multipoles[0] = (1 + 2/3*beta + 1/5*beta**2) * b1**2 * Plin(k)
        model_multipoles[1] = (4/3*beta + 4/7*beta**2) * b1**2 * Plin(k)

    D_M = (data_multipoles - model_multipoles).flatten()
    return D_M@(C_inv @ D_M)
```

```python
kmax_range = possible_kmax[:11]
Nkmax = len(kmax_range)
header = 'parameter guess for each density bin at kmax=%s'%str(kmax_range)

for damped in [False, True]:
    print('Damped: %s'%damped)
    b1_guess, beta_guess, sigma_guess = np.full((n_ptile, Nkmax), np.nan), np.full((n_ptile, Nkmax), np.nan), np.full((n_ptile, Nkmax), np.nan)
    for i in range(n_ptile):
        t1 = time.time()
        for j,kmax in enumerate(kmax_range):
            mask = np.full(len(k_full), False)
            mask = k_full <= kmax
            mask[0] = False 
            k_sliced = k_full[mask]
            Pk_ells_i = Pk_ells_full[:,:,mask][i]
            cov_mat = np.loadtxt('/home/jwack/main/bruteforce_covmat/covariance_matricies/cov_ptile_%d.txt'%i)
            C_inv = icf.mock_cov_mat_inv(cov_mat, k_full, kmax)
            
            if damped:
                start = [1,1,3]
                bounds = ((None, None), (None, None), (None, None))
            else:
                start = [1,1]
                bounds = ((None, None), (None, None))
                
            s = minimize(chi2, start, args=(Pk_ells_i, k_sliced, C_inv, damped), bounds=bounds)
            b1_guess[i][j] = s.x[0]
            beta_guess[i][j] = s.x[1]
            if damped:
                sigma_guess[i][j] = s.x[2]
                
        t2 = time.time()
        print("Fitted %d-th percentile in %s"%(i+1,str(timedelta(seconds=t2-t1))))   
        
    if damped:
        np.savetxt('fit_prep/b1_guess_damped.txt', b1_guess, header=header)
        np.savetxt('fit_prep/beta_guess_damped.txt', beta_guess, header=header)
        np.savetxt('fit_prep/sigma_guess_damped.txt', sigma_guess, header=header)
    else:
        np.savetxt('fit_prep/b1_guess.txt', b1_guess, header=header)
        np.savetxt('fit_prep/beta_guess.txt', beta_guess, header=header)
    print('-'*32)
```

```python
fig, axs = plt.subplots(1,3,figsize=(26,12))

b1_guess = np.loadtxt('fit_prep/b1_guess.txt')
beta_guess = np.loadtxt('fit_prep/beta_guess.txt')

b1_damped_guess = np.loadtxt('fit_prep/b1_guess_damped.txt')
beta_damped_guess = np.loadtxt('fit_prep/beta_guess_damped.txt')
sigma_guess = np.loadtxt('fit_prep/sigma_guess_damped.txt')

for i in range(n_ptile):
    line, = axs[0].plot(kmax_range, b1_damped_guess[i], label='%d-th'%i)
    axs[0].plot(kmax_range, b1_guess[i], c=line.get_color(), alpha=0.5)
    line, = axs[1].plot(kmax_range, beta_damped_guess[i], label='%d-th'%i) 
    axs[1].plot(kmax_range, beta_guess[i], c=line.get_color(), alpha=0.5) 
    axs[2].plot(kmax_range, sigma_guess[i], label='%d-th'%i)    
    
axs[0].set_title('$b_1$ guess')
axs[0].set_xlabel(r'$kmax$ [$h \ \mathrm{Mpc}^{-1}$]')
axs[0].set_ylabel(r'$b_1$')

axs[1].set_title(r'$\beta$ guess')
axs[1].set_xlabel(r'$kmax$ [$h \ \mathrm{Mpc}^{-1}$]')
axs[1].set_ylabel(r'$\beta$')

axs[2].set_title(r'$\sigma$ guess')
axs[2].set_xlabel(r'$kmax$ [$h \ \mathrm{Mpc}^{-1}$]')
axs[2].set_ylabel(r'$\sigma$')

handles, labels = plt.gca().get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, 0.01), ncol=n_ptile)
plt.show()

#fig.savefig("plots/parameter_guesses.pdf")
```

### Interpretation
The pure Kaiser model is shown in transparent lines while Kaiser+FoG is plotted with solid lines. 
At $k>0.08$ both models seem to lead to very similar guesses of the parameter values. For $k<0.08$ the damped model predicts a smaller $b_1$ but a larger $\beta$. This anti correlation is expected to assure a constant $f_0$.

Note that for the damped model, the guesses for the smallest kmax are close to the starting positions passed to `scipy.optimize.minimize`.

For $\sigma$ no clear trend visible other than for all bins and all kmax, the parameter guess seems to be clustered around the starting position. Hence, one should not use the computed values as a reliable estimate of the true mean value. Below one can plot a polynomial interpolation for $\sigma$.

```python
for i in range(n_ptile):
    line, = plt.plot(kmax_range, np.abs(b1_damped_guess[i]*beta_damped_guess[i]), label='%d-th'%i)
    plt.plot(kmax_range, np.abs(b1_guess[i]*beta_guess[i]), c=line.get_color(), alpha=0.5)
    
plt.title('$f_0$ guess')
plt.xlabel(r'$kmax$ [$h \ \mathrm{Mpc}^{-1}$]')
plt.ylabel(r'$f_0$')

handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=5)
plt.show()
```

Note that for some bins $f_0$ would be negative. By using the symmetry of the parameter space about $b_1=0$ we can enforce $f_0>0$.

```python
poly_degree = int(1)

for i in range(n_ptile):
    trend = np.poly1d(np.polyfit(kmax_range, sigma_guess[i], poly_degree))
    line, = plt.plot(kmax_range, sigma_guess[i], label='%d-th'%i, alpha=0.3)
    plt.plot(kmax_range, trend(kmax_range), c=line.get_color(), linestyle='--')
    
plt.title('$\sigma$ guess interpolated')
plt.xlabel(r'$kmax$ [$h \ \mathrm{Mpc}^{-1}$]')
plt.ylabel(r'$\sigma$')

handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, labels, loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=5)
plt.show()
```
