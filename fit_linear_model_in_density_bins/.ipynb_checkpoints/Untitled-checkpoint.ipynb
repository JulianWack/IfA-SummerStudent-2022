```python
from classy import Class

import numpy as np
import matplotlib.pyplot as plt
import time
from scipy.stats import bayes_mvs

import inv_cov_funcs as icf
import classylss
import fitsio
import zeus 
from nbodykit.lab import *
from nbodykit import style, setup_logging
plt.style.use(style.notebook)
```

```python
z = 0.61

Planck18_settings = {'A_s':2.083e-9,
                     'n_s':0.9649,
                     'tau_reio':0.0544,
                     'omega_b':0.02237,
                     'omega_cdm':0.12,
                     'h':0.6736,
                     'N_ur':2.0328,
                     'N_ncdm':1,
                     'alpha_s':0.0,
                     #'omega_ncdm':0.0006442, # causes CosmoComputationError
                   # other output and precision parameters
                     #'P_k_max_1/Mpc':100.0,
                     'z_pk':z}  



M = Class()
M.set(Planck18_settings)

M.set({'output':'mPk',
       'non linear':'PT',
       'IR resummation':'Yes',
       'Bias tracers':'Yes',
       'cb':'Yes',
       'RSD':'Yes',
       'AP':'Yes',
       'Omfid':'0.31'
      })
M.compute()
```

```python
##### NUISANCE PARAMETERS #### must be fitted in analysis
b1 = 2.0
cs = 1. # in units [Mpc/h]^2
b2 = -1.
bG2 = 0.1
bGamma3 = -0.1
Pshot = 5e3 # in units [Mpc/h]^3
cs0 = 5. # in units [Mpc/h]^2
cs2 = 15. # in units [Mpc/h]^2
cs4 = -5. # in units [Mpc/h]^2
b4 = 100. # in units [Mpc/h]^4
##############################

h = M.h()
f0_true = M.scale_independent_growth_factor_f(z)
kvec = np.logspace(-3,np.log10(3),1000) # array of kvec in h/Mpc
khvec = kvec*h # array of kvec in 1/Mpc

M.initialize_output(khvec, z, len(khvec))
pk_g0 = M.pk_gg_l0(b1, b2, bG2, bGamma3, cs0, Pshot, b4)
pk_g2 = M.pk_gg_l2(b1, b2, bG2, bGamma3, cs2, b4)
pk_g4 = M.pk_gg_l4(b1, b2, bG2, bGamma3, cs4, b4)
```

```python
plt.plot(kvec,np.array(pk_g0)*kvec,color='darkgreen',linestyle='-',label=r'$\ell = 0$')
plt.plot(kvec,np.array(pk_g2)*kvec,color='darkorange',linestyle='-',label=r'$\ell = 2$')
plt.plot(kvec,np.array(pk_g4)*kvec,color='purple',linestyle='-',label=r'$\ell = 4$')

plt.xlim([1e-3,0.3])
plt.ylim([1,2500])
```

```python
cosmo = cosmology.cosmology.Cosmology.from_dict(Planck18_settings)
Plin = cosmology.LinearPower(cosmo, z, transfer='EisensteinHu')
```

```python
pk_lin_old = Plin(kvec)
pk_lin_new = np.asarray([M.pk_lin(kh,z)*h**3. for kh in khvec])

plt.plot(kvec, np.abs(pk_lin_old-pk_lin_new))
plt.yscale('log')
```

```python
beta=f0_true/b1
ell_0 = (1 + 2/3*beta + 1/5*beta**2) * b1**2 * pk_lin_old + Pshot
ell_2 = (4/3*beta + 4/7*beta**2) * b1**2 * pk_lin_old
```

```python
plt.plot(kvec,pk_g0-ell_0,color='darkgreen',linestyle='-',label=r'$\ell = 0$')
plt.plot(kvec,pk_g2-ell_2,color='darkorange',linestyle='-',label=r'$\ell = 2$')
plt.xlim([1e-3,0.3])
plt.ylim([-2e3,5e2])
```
