# Comparing s space power spectrum to linear theory and a simple non-linear correction
Get linear theory s space power spectrum from data and compare to Kaiser's linear theory result. Also compare data against simple correction of Kaiser's result to account for (some) non-linearity. For the relation between real and redshift space power, consult Kaiser (1987) eq. $(3.5)$ and/or Hamilton (1997) eq. $(4.36)$. Note that the power spectrum is isotropic in real space but anisotropic (direction dependence through $\mu$) in redshift space.

```python
Nmesh = 256
k_Nyq = np.pi*Nmesh/BoxSize
k_min = 0.005
k_max = k_Nyq*1.05 # FFTPower only computes upto k_max 
Nk = int((k_max - k_min)/delta_k) # number of bins

kbin = [k_min,k_max,Nk,'lin']
```

```python
cat = nbodypk.make_cat(filename, cosmo=cosmo, LOS=LOS, z=redshift)
mesh = cat.to_mesh(position='RSDPosition', resampler='tsc', BoxSize=BoxSize, Nmesh=Nmesh, compensated=True)
```

```python
%time Pkmu, mus = nbodypk.get_binned_Pkmu(mesh, 5, LOS, kbin)
```

```python
def kaiser_pkmu(k, mu, cosmo, redshift, b1):
    """Returns power spectrum in redshift space, following Kaiser's linear results.
    Requires linear theory r space power spectrum"""
    Plin = cosmology.LinearPower(cosmo, redshift, transfer='EisensteinHu')
    beta = cosmo.scale_independent_growth_rate(redshift) / b1 
    
    return (1 + beta*mu**2)**2 *b1**2 * Plin(k)

def non_linear_kaiser_pkmu(k, mu, cosmo, redshift, b1):
    '''Uses phenomenological model to account for non-linearity. The value of sigma is conventional.'''
    sigma = np.sqrt(2) # km/s
    
    return np.exp(-0.5*(sigma*k*mu)**2)*kaiser_pkmu(k, mu, cosmo, redshift, b1)
    
```

Above introduces the linear theory prediction by Kaiser and a phenomenological model to account for non-linearity by convolving the linear correlation function with a random isotropic pairwise velocity dispersion function $f(v)$. By the convolution theorem, this results in multiplying Kaiser's result by $FT[f(v)] \equiv \hat{f}(k \mu)$. Conventional choices of $f(v)$ and its FT are given in Hamilton's review (eqs. (7.3), (7.4)) where $\sigma$ gives the 1D pairwise velocity dispersion which is conventionally taken $\sqrt{2}$ times larger than a typical random 1D galaxy dispersion. See Hamilton's review section 7.2 for references. Usually sigma of order $500 \frac{km}{s}$. This is found much too large as this essentially forces the exp to zero, such that the non-linear Kaiser expression vanishes as well. Here we use a Gaussian $f(v)$ whose FT is also Gaussian.

```python
fig = plt.figure()
colours = ['blue', 'orange', 'green', 'red', 'purple']
ks = Pkmu[0][:,0] # ks the same for all mu
for i,mu in enumerate(mus):
    Pk = Pkmu[i]
    label = r'$\mu$=%.1f' %mu
    plt.plot(ks, Pk[:,1]/kaiser_pkmu(ks, mu, cosmo, redshift, b1), alpha=0.3, c=colours[i])
    plt.plot(ks, Pk[:,1]/non_linear_kaiser_pkmu(ks, mu, cosmo, redshift, b1), label=label, c=colours[i])

plt.axhline(y=1, color='grey', linestyle='--')

plt.title("Comparison to (non-) linear theory in redshift space")
plt.xlabel(r"$k$ [$h \ \mathrm{Mpc}^{-1}$]")
plt.ylabel(r"$P(k, \mu) / P_\mathrm{Kaiser}(k,\mu)$")
plt.legend(ncol=2, frameon=False)
plt.show()

fname = "plots/s_space_nonlinear_comparison.pdf"
#fig.savefig(fname)
```

### Interpretation
- Regardless of direction relative to LOS, there always exists a maximum $k$ up to which the computed power spectrum agrees with linear theory. However, for directions further away from the LOS, this value of $k$ becomes smaller.
- Large $|\mu|$ gets suppressed more strongly for large $k$ and damped below Kaiser's prediction. This damping appears to be partially resolved when introducing the phenomenological model non-linearity model, albeit with an unrealistic guess for $\sigma$. In fact, it appears that this causes all $\mu$ to overestimate the non-linear Kaiser result.
- $\mu=0$ over estimates linear theory drastically. This remains unaffected by the non-linearity correction as $\hat{f}(k \mu) = 1$ for $\mu=0$. 
- Power spectrum approximately even in $\mu$: Plots for $\pm \mu$ almost identical
- All spectra drop to 0 at $k=k_{Nyq}$ as P(k,mu) = 0 due to binning 

# Estimating $\beta$:
Under the assumption that the data is based on the 'Planck15' cosmology, the true value of $\beta$ can be computed. In the following two methods are considered which allow to estimate $\beta$ from the mock catalog.

```python
beta_true = cosmo.scale_independent_growth_rate(redshift) / b1
print("True value of beta: {:.2f}".format(beta_true))
```

## A: Ratio of quadrupole to monopole
Following Section $5.2$ of Hamilton's review: Kaiser's linear result predicts that the redshift space power spectrum is given by the sum of a monopole ($O(\mu^0)$), quadrupole ($O(\mu^2)$), and hexadecapole ($O(\mu^4)$) times the real space power spectrum. Upon decomposing the redshift space power spectrum into harmonics (become Legendre polynomials due to symmetry) the expansion coefficients can be identified (see eq $(5.7)$). Specifically:
$$\frac{P_2^s(k)}{P_0^s(k)} = \frac{4/3 \beta + 4/7 \beta^2}{1 + 2/3 \beta + 1/5 \beta^2}$$
Now find LHS for given catalog and fit constant line to get the value of $\beta$.

When fitting a higher order polynomial rather than just a constant, I regard these as non-linear corrections (as they introduce $k$ dependence). For all cases, $\beta$ will be found from the constant term in the fitted polynomial. 

Similar computations as in power spectrum of simulation box Cookbook

```python
%time r = FFTPower(mesh, mode='2d', dk=0.005, kmin=0.01, Nmu=5, los=LOS, poles=[0,2,4]).poles
k = r['k']
# subtract shotnoise only from monopole and discard complex part for all poles
P_0 = r['power_0'].real - r.attrs['shotnoise'] 
P_2 = r['power_2'].real
P_4 = r['power_4'].real
```

```python
ratio = P_2/P_0

degree = 1
fit_paras = np.polyfit(k, ratio, degree)

# Solve for beta given best fit value of LHS ratio. Pass coefficients of quadratic in beta
coeff = [1/5*fit_paras[-1]-4/7, 2/3*fit_paras[-1]-4/3, fit_paras[-1]]
beta = np.roots(coeff)[-1] # first root always negative and need 0<beta<1
```

```python
def make_poly_str(fit_paras):
    '''Make strings of fitted polynominal for plot.'''
    poly_str = '${:.2f}'.format(fit_paras[-1])
    for i,co in enumerate(np.flip(fit_paras)):
        if i == 0:
            continue
        if co > 0:
            poly_str += '+{:.2f}k^{:d}'.format(co,i)
        elif co < 0:
            poly_str += '{:.2f}k^{:d}'.format(co,i)

    return poly_str + '$'
```

```python
fig = plt.figure()

plt.plot(k, ratio)
plt.plot(k, np.poly1d(fit_paras)(k), label=make_poly_str(fit_paras))

plt.title("Ratio of quadrupole to monopole of s-space power spectrum")
plt.xlabel(r"$k$ [$h \ \mathrm{Mpc}^{-1}$]")
plt.ylabel(r"$P_2(k)/P_0(k)$")
plt.text(0.71, 0.86, r"$\beta={:.2f}, Nmesh={:d}$".format(beta, Nmesh), transform=plt.gca().transAxes)
plt.legend(frameon=False)

plt.show()
fname = "plots/beta_estimate_linear.pdf"
#fig.savefig(fname)
```

```python
print("Error: {:.3%}".format(np.abs(1-beta/beta_true)))
```

### Interpretation
The pure linear theory result of of obtaining a constant ratio of quadrupole to monopole fails. Fitting a constant produces a physically unreasonable $\beta <0$. Thus I consider a linear fit as the next highest polynomial. I interpret the k dependent term of the 1st order polynomial as the first order non-linear correction, while I assume the constant term to still carry its interpretation from the linear theory which allows me to estimate $\beta$. For $Nmesh=64,512$ get $\beta \approx 0.54$. This result appears to be in the same ball park as other observations (See Hamilton's review, table 1 on p.65). However, for $Nmesh=1024$ the linear fit doesn't appear suitable (for large $k$ the ratio stabilizes but the oscillations become large too). A cubic fit produces $\beta=0.64$. 

Note that a cubic fit will also work for the samller choices of Nmesh: more parameter mean an easier fit!

## B: Ratio of angle averaged redshift space to real space power spectrum

Following section $(5.1)$ of Hamilton's review. Note that the angle averaged redshift space power spectrum is just $P_0^s(k)$. Hence, this estimation method is just using the coefficients from the expansion of $P^s(\vec{k})$ in terms of harmonics. Specifically, the coefficient of the zero-th order Legendre polynomial is used: $P_0^s(k) = (1 + \frac{2}{3} \beta + \frac{1}{5} \beta^2)P_{lin}(k) = \alpha(\beta) P_{lin}(k)$. See eq $(5.1), (5.11)$. Note hat $\alpha \in [1,28/15]$ as $\beta \in [0,1]$.

Thus the two presented methods of estimating $\beta$ can be considered to be almost identical. Note that $P_{lin}(k)$ refers to the linear power spectrum in real space.

```python
Plin = cosmology.LinearPower(cosmo, redshift, transfer='EisensteinHu')
ratio = P_0/Plin(k)

degree = 1
fit_paras = np.polyfit(k, ratio, degree)

coeff = [1/5, 2/3, 1-fit_paras[-1]]
beta = np.roots(coeff)[-1]
```

```python
fig = plt.figure()

plt.plot(k, ratio)
plt.plot(k, np.poly1d(fit_paras)(k), label=make_poly_str(fit_paras))

plt.title("Ratio of monopole to r-space power spectrum")
plt.xlabel(r"$k$ [$h \ \mathrm{Mpc}^{-1}$]")
plt.ylabel(r"$P_0(k)/P_{lin}(k)$")
plt.text(0.02, 0.86, r"$\beta={:.2f}, Nmesh={:d}$".format(beta, Nmesh), transform=plt.gca().transAxes)
plt.legend(frameon=False)

plt.show()
fname = "plots/beta_estimate_linear_2.pdf"
#fig.savefig(fname)
```

### Interpretation
As in the first method to estimate $\beta$, the purely linear theory result of a constant ratio fails. There isn't even a small $k$ region in which the ratio appears to be roughly constant. For a linear fit wit $Nmesh=512, 1024$ I obtain $\beta=0.57, 0.61$ both of which seem reasonable. A cubic fit for $Nmesh=1024$ gives $\beta=0.45$ but appear to be not required as the linear fit mimics the ratio well.

## Smoothing window function
We expect the data to contain a linear and a non-linear part. To only test the results of Kaiser, we want to only consider the linear part of the data. This can be done by smoothing out the power spectrum over small $k$ through a window function (see Hamilton's review eq $(5.2), (5.11)$):
$$ 
\tilde{P^s} \equiv \int_0^{\infty} 4\pi W(k) P^s(k) k^2 \frac{dk}{(2\pi)^3}, \text{and analogous for }P^s_l(k).
$$ 
Option for normalized window function ($\int_0^{\infty} W(k) \frac{d^3k}{(2\pi)^3} = 1$) is $$ \frac{W(k)}{(2\pi)^3} = \frac{32\pi^2}{\Gamma(\frac{n+3}{2})} k^n e^{-k^2}$$ for even integer n. 

Note that the smoothing removes the $k$ dependence of the ratios of interest: $\frac{\tilde{P_2^s}(k)}{\tilde{P_0^s}(k)}$ and $\frac{\tilde{P_0^s}(k)}{\tilde{P_{lin}}(k)}$ (below referred to as $A$, $B$ respectively).

```python
from scipy.special import gamma
from scipy.integrate import simpson

def smooth_power(k, power, n):
    '''Smooths out power as explained in markdown. Not that integration is only possible upto k_Nyq'''
    integrand = 128*np.pi**3 / gamma((n+3)/2) * k**(n+2) * np.exp(-k**2) * power
    return simpson(integrand, k)
```

```python
ns = np.arange(-4, 6, 2)

print("{:<8} {:<15} {:<10}".format('n','beta_A','beta_B'))
print("-"*32)
for n in ns:
    # A: quadrupole to monopole
    ratio_A = smooth_power(k, P_2, n) / smooth_power(k, P_0, n)
    coeff_A = [1/5*ratio_A-4/7, 2/3*ratio_A-4/3, ratio_A]
    beta_A = np.roots(coeff_A)[-1]

    # B: monopole to real space
    ratio_B = smooth_power(k, P_0, n) / smooth_power(k, Plin(k), n)
    coeff_B = [1/5, 2/3, 1-ratio_B]
    beta_B = np.roots(coeff_B)[-1]
    print("{:<8} {:<15.2f} {:<10.2f}".format(n, beta_A, beta_B))
    print("-"*32)
```

### Interpretation
Only acceptable values of $\beta$ obtained for $n=-2,-4$ corresponding to accepting power up to $k \approx 2$ and $k \approx 0.2$ (for Nmesh=1024). Larger $n$ causes peak of window to be shifted to larger $k$ which we aim to ignore in the first place. Below I plotted the functional form of the filter function $G(k)$ for $\tilde{P}^s = \int G(k) P^s(k) dk$ i.e. $G(k) \sim k^2 W(k)$.

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(24,8))

ax1.plot(k, P_0, label="$P^s_0(k)$")
ax1.plot(k, P_2, label="$P^s_2(k)$")
ax1.plot(k, Plin(k), label="$P_{lin}(k)$")
ax1.set_title("Power")
ax1.set_xlabel('k')
ax1.set_yscale('log')
ax1.legend(frameon=False)

ax2.plot(k, k**(-2+2) * np.exp(-k**2))
ax2.set_title('Filter function shape: $n=-2$')
ax2.set_xlabel('$k$')
ax2.set_ylabel(r'$G(k) \sim k^2W(k)$')

ax3.plot(k, k**(-4+2) * np.exp(-k**2))
ax3.set_title('Filter function shape: $n=-4$')
ax3.set_xlabel('$k$')
ax3.set_ylabel(r'$G(k) \sim k^2W(k)$')

plt.show()
```

## Non-linearity via Bromley, Warren, Zurek 1997
Simple process to account for non-linearity summarized in Hamilton's review section $(8.2.3)$:
$$
\tilde{P^s_0} \to \tilde{P^s_0} + \sigma^2(\frac{1}{6}\tilde{P^s_0}^{NL} + \frac{1}{15}\tilde{P^s_2}^{NL}) 
$$

$$
\tilde{P^s_2} \to \tilde{P^s_2} + \sigma^2(\frac{1}{3}\tilde{P^s_0}^{NL} + \frac{11}{42}\tilde{P^s_2}^{NL} + \frac{2}{21}\tilde{P^s_4}^{NL}) 
$$

with $\tilde{P^s_l}^{NL} \equiv \int_0^{\infty} 4\pi W(k) P^s_l(k) k^4 \frac{dk}{(2\pi)^3}$ i.e only additional power of $2$ compared to normal smoothing via window. As $n=-2$ gave best results earlier, use `smooth_power(n=0)` on $P^s_l(k)$ to get $\tilde{P^s_l}^{NL}$.

Authors used $\sigma=320$ km/s.

```python
sigma = 1 # km/s
n = -2

P_0_NL = smooth_power(k,P_0,n+2)
P_2_NL = smooth_power(k,P_2,n+2)
P_4_NL = smooth_power(k,P_4,n+2)

new_P_0_tilde = smooth_power(k, P_0, n=n) + sigma**2* (1/6*P_0_NL + 1/15*P_2_NL)
new_P_2_tilde = smooth_power(k, P_2, n=n) + sigma**2* (1/3*P_0_NL + 11/42*P_2_NL + 2/21*P_4_NL)

# A: quadrupole to monopole
ratio_A = new_P_2_tilde / new_P_0_tilde
coeff_A = [1/5*ratio_A-4/7, 2/3*ratio_A-4/3, ratio_A]
beta_A = np.roots(coeff_A)[-1]
print("Quadrupole to monopole ratio: beta = {:.2f}".format(beta_A))

# B: monopole to real space
ratio_B = new_P_0_tilde / smooth_power(k, Plin(k), n)
coeff_B = [1/5, 2/3, 1-ratio_B]
beta_B = np.roots(coeff_B)[-1]
print("Monopole to real space ratio: beta = {:.2f}".format(beta_B))
```

### Interpretation 
Results more similar to ones obtained without using a window function and fitting linear/cubic. However, value of $\sigma$ not realistic.
